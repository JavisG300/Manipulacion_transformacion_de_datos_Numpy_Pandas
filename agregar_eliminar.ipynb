# ¿Cómo eliminar datos en Pandas?

## Eliminar columnas

```python
df_books.drop('Genre',axis = 1).head(2) #Recordando 0 filas 1 columnas
```

¿Pero y si ejecuto el dataframe original?

```python
df_books.head(2)
```

**drop** borra de la salida, no directamente del df

Para borrar en el df orignal se usa **inplace = True**

```python
df_books
```

```python
df_books.drop('Genre',axis = 1, inplace = True) #Recordando 0 filas 1 columnas
```

Ahora si se borró del df original

```python
df_books.head()
```

Otra forma de lograr el borrado

```python
df_books = df_books.drop('Year', axis = 1)
```

```python
df_books
```

```python
del df_books['Price'] #Easta funcion no es propia de pandas, es de python
```

```python
df_books
```

## Borrar filas

```python
df_books.drop(0, axis = 0)
```

```python
df_books.drop([0,1,2], axis = 0).head()
```

```python
df_books.drop(range(0,10), axis = 0).head() #El 10 no lo borra. Sino usa inplace no se guarda en el df original
```
